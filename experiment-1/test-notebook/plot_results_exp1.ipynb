### stacking

```python
plt.figure(figsize=(5,3))
b1 = plt.bar(x,y0,color="yellow")
b1[4].set_color('r')
b1[5].set_color('r')
b2 = plt.bar(x,y1,color="green")
b1[4].set_color('b')
b1[5].set_color('b')
plt.xlabel("Model", fontsize=12)
plt.ylabel("F1 score", fontsize=12)
plt.title("F1 class 0 vs class 1", fontsize=12)
plt.show()
```
