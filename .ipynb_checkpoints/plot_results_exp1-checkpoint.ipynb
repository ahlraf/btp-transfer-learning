```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ["BERT/A", "BERT/F1", "BERT/R", "BERT/P", "NB", "SVM"]
y0 = [0.934, 0.92, 0.916, 0.924, 0.794, 0.734]
y1 = [0.932, 0.93, 0.896, 0.922, 0.702, 0.74]
y2 = [0.932, 0.926, 0.906, 0.926, 0.748, 0.738]
```

```python
x = ["BERT/A", "BERT/F1", "BERT/R", "BERT/P", "NB", "SVM"]
y0 = [0.932, 0.932, 0.888, 0.924, 0.728, 0.738]
y1 = [0.934, 0.918, 0.922, 0.922, 0.774, 0.736]
y2 = [0.932, 0.926, 0.904, 0.926, 0.752, 0.738]
```

```python
plt.figure(figsize=(5,3))
b1 = plt.bar(x,y0,color="green")
b1[4].set_color('r')
b1[5].set_color('r')
plt.xlabel("Model", fontsize=12)
plt.ylabel("Recall (class 0)", fontsize=12)
plt.title("Recall (class 0)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(5,3))
b1 = plt.bar(x,y1,color="green")
b1[4].set_color('r')
b1[5].set_color('r')
plt.xlabel("Model", fontsize=12)
plt.ylabel("Recall (class 1)", fontsize=12)
plt.title("Recall (class 1)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(5,3))
b1 = plt.bar(x,y2,color="green")
b1[4].set_color('r')
b1[5].set_color('r')
plt.xlabel("Model", fontsize=12)
plt.ylabel("Recall (macro-avg)", fontsize=12)
plt.title("Recall (macro-avg)", fontsize=12)
plt.show()
```
